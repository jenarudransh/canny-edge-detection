```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img=cv2.imread(r"C:\SEM 6 (ECE)\Robotic Automation in ROS and C++\Experiments\ReadDisplaySaveImage\color_parrot.jpg")
rgb=cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Multiply by scaler value of 70, matrix has dimention same as input image
intensity_matrix=np.ones(img.shape,dtype='uint8')*70

# substract intensity matrix from input image in order to dectrase the brightness
bright_img=cv2.subtract(img, intensity_matrix)
rgb2=cv2.cvtColor(bright_img, cv2.COLOR_BGR2RGB)

plt.subplot(1,2,1), plt.imshow(rgb), plt.title('Original Image')
plt.subplot(1,2,2), plt.imshow(rgb2), plt.title('Bright Image')
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img=cv2.imread(r"C:\SEM 6 (ECE)\Robotic Automation in ROS and C++\Experiments\ReadDisplaySaveImage\color_parrot.jpg",0)
height,width=img.shape
size=img.size
print(height,width)
s = np.int64(0)  # Use int64 to prevent overflow
for i in range(height):
    for j in range(width):
        s = s + img[i][j]
        
mean_s=s/(height*width)

s = np.float64(0)  # Use float64 for variance calculation to prevent overflow
for i in range(height):
    for j in range(width):
        s = s + (np.float64(img[i][j]) - mean_s) ** 2

var_s=s/(height*width)
sdev=var_s**0.5
print("cintrast of an gray Image", sdev)
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img=cv2.imread(r"C:\SEM 6 (ECE)\Robotic Automation in ROS and C++\Experiments\ReadDisplaySaveImage\color_parrot.jpg",0)
# Note: Ensure height and width are defined if not running previous cells
height, width = img.shape 
p1=np.zeros((256,1))

for i in range(height):
    for j in range(width):
        p1[img[i][j]] = p1[img[i][j]] + np.int64(1)  # Use int64 for counting

p=p1/p1.sum()
s = np.float64(0)  # Use float64 for entropy calculation
for i in range(256):
    if p[i]>0:
        s=s+(p[i]*np.log2(p[i]))

entropy_value= -s
plt.imshow(img)
plt.title(entropy_value)
plt.show()
```
