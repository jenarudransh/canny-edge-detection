```python
# Convert image into grayscale

import cv2
image = cv2.imread("C:\\Users\\Rudransh\\Downloads\\wallpaper.jpg")
image = cv2.resize(image,(700,600))
cv2.imshow("original", image)
cv2.waitKey(0)

# Use the cvtColor() function to grayscale the image
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
cv2.imshow("Grayscale", gray_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Brightness of gray-scale image
import numpy as np
mean_z = np.mean(gray_image)
print("Brightness of gray-scale image:", mean_z)

mean_o = np.mean(image)
print("Brightness of original image:", mean_o)
```

```python
import cv2
import numpy as np

image = cv2.imread("C:\\Users\\Rudransh\\Downloads\\wallpaper.jpg")
image = cv2.resize(image,(700,600))
# Define a brightness value. A positive value increases brightness,
# while a negative value decreases it.
brightness_value = 50

# Increase brightness by adding the brightness value to all pixels
bright_image = cv2.convertScaleAbs(image, alpha=1, beta=brightness_value)

cv2.imshow("Original Image", image)
cv2.imshow("Brightened Image", bright_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
image = cv2.imread("C:\\Users\\Rudransh\\Downloads\\wallpaper.jpg")
image = cv2.resize(image,(700,600))

# Define a brightness value. A positive value increases brightness,
# while a negative value decreases it.
brightness_value = -50

# Decrease brightness by subtracting the brightness value from all pixels
dark_image = cv2.convertScaleAbs(image, alpha=1, beta=brightness_value)

cv2.imshow("Original Image", image)
cv2.imshow("Darkened Image", dark_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
image = cv2.imread("C:\\Users\\Rudrash\\Downloads\\wallpaper.jpg")
image = cv2.resize(image,(400,300))

# Define contrast increase and decrease factors
alpha_increase = 1.5   # Increase contrast
alpha_decrease = 0.5   # Decrease contrast

# Adjust contrast by multiplying each pixel by the contrast value (alpha)
increased_contrast = cv2.convertScaleAbs(image, alpha=alpha_increase, beta=0)
decreased_contrast = cv2.convertScaleAbs(image, alpha=alpha_decrease, beta=0)

cv2.imshow("Original Image", image)
cv2.imshow("Increased Contrast Image", increased_contrast)
cv2.imshow("Decreased Contrast Image", decreased_contrast)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
from scipy.stats import entropy
import matplotlib.pyplot as plt
import cv2
import numpy as np

image = cv2.imread("C:\\Users\\SK\\Desktop\\tomato.jpg", cv2.IMREAD_GRAYSCALE)
image = cv2.resize(image, (700, 600))

# Display the image using matplotlib
plt.imshow(image, cmap='gray')
plt.title('Grayscale Image')
plt.axis('off')
plt.show()

# Calculate the histogram of the image
hist, _ = np.histogram(image, bins=256, range=(0, 256))

# Normalize the histogram to get the probability distribution
hist_norm = hist / hist.sum()

# Calculate entropy
image_entropy = entropy(hist_norm, base=2)

print(f'Entropy of the image: {image_entropy}')
```

```python
from scipy.stats import skew
import cv2
import numpy as np
import matplotlib.pyplot as plt

image = cv2.imread("C:\\Users\\SK\\Desktop\\tomato.jpg", cv2.IMREAD_GRAYSCALE)
image = cv2.resize(image, (700, 600))

# Calculate the skewness of the original image
original_skewness = skew(image.flatten())
print(f'Skewness of the original image: {original_skewness}')

# Create an artificial skew by applying a power transform to the image
gamma = 2.0  # Gamma value > 1 will darken the image, < 1 will lighten the image (artificial skew)
skewed_image = np.power(image / 255.0, gamma) * 255.0
skewed_image = skewed_image.astype(np.uint8)

# Calculate the skewness of the skewed image
skewed_skewness = skew(skewed_image.flatten())
print(f'Skewness of the skewed image: {skewed_skewness}')

# Display the original and skewed images
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.imshow(image, cmap='gray')
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(skewed_image, cmap='gray')
plt.title('Skewed Image')
plt.axis('off')

plt.show()
```

```python
# Original image
plt.subplot(2, 2, 1)
plt.imshow(image, cmap='gray')
plt.title(f'Original Image\nSkewness: {original_skewness:.4f}')
plt.axis('off')

# Skewed image
plt.subplot(2, 2, 2)
plt.imshow(skewed_image, cmap='gray')
plt.title(f'Skewed Image\nSkewness: {skewed_skewness:.4f}')
plt.axis('off')

# Histogram of original image
plt.subplot(2, 2, 3)
plt.hist(image.flatten(), bins=256, color='blue', alpha=0.7)
plt.title('Original Image Histogram')
plt.xlabel('Pixel Intensity')
plt.ylabel('Frequency')

# Histogram of skewed image
plt.subplot(2, 2, 4)
plt.hist(skewed_image.flatten(), bins=256, color='red', alpha=0.7)
plt.title('Skewed Image Histogram')
plt.xlabel('Pixel Intensity')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
from scipy.stats import kurtosis

image = cv2.imread("C:\\Users\\Rudransh\\Downloads\\wallpaper.jpg", cv2.IMREAD_GRAYSCALE)
image = cv2.resize(image, (700, 600))

# Flatten the image array to 1D
pixels = image.flatten()

# Calculate the kurtosis of the pixel values
image_kurtosis = kurtosis(pixels)

print(f'Kurtosis of the image: {image_kurtosis}')

plt.figure(figsize=(10, 5))
```

```python
# Original image display
plt.subplot(1, 2, 1)
plt.imshow(image, cmap='gray')
plt.title('Grayscale Image')
plt.axis('off')

# Histogram with kurtosis value
plt.subplot(1, 2, 2)
plt.hist(pixels, bins=256, color='green', alpha=0.7)
plt.title(f'Image Histogram\nKurtosis: {image_kurtosis:.4f}')
plt.xlabel('Pixel Intensity')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```
